```python
# Un peu comme une conductance
def weightingRatio(G, gt_partition):
    edges=G.edges()
    poidsIntra=0
    poidsInter=0
    for (u,v) in edges:
        if gt_partition.subsetOf(u) == gt_partition.subsetOf(v):
            poidsIntra+=G.weight(u,v)
        else:
            poidsInter+=G.weight(u,v)
    return (poidsInter/poidsIntra)
    
```

```python
def ratios(G, gt_partition):
    edges=G.edges()
    poidsIntra=0
    poidsInter=0
    nbInter=0
    nbIntra=0
    for (u,v) in edges:
        if gt_partition.subsetOf(u) == gt_partition.subsetOf(v):
            poidsIntra+=G.weight(u,v)
            nbIntra+=1
        else:
            poidsInter+=G.weight(u,v)
            nbInter+=1
    return (nbInter,nbIntra, poidsInter, poidsIntra)
    
```

```python
def node_recall(G, gt_partition):
    nodes=G.nodes()
    deg_nodes=dict()
    deg_out_nodes=dict()
    for n in nodes:
        deg_nodes[n]=0
        deg_out_nodes[n]=0
    edges=G.edges()
    for (u,v) in edges:
        deg_nodes[u]+=G.weight(u,v)
        deg_nodes[v]+=G.weight(u,v)
        if gt_partition.subsetOf(u) != gt_partition.subsetOf(v):
            deg_out_nodes[u]+=G.weight(u,v)
            deg_out_nodes[v]+=G.weight(u,v)
    return (deg_nodes, deg_out_nodes)
```

```python
chemin1="lfr_5000"
import os
gpath="network.dat"
cpath="community.dat"
mks=os.listdir(chemin1)
import networkit as nk
from scipy.stats import spearmanr
# %%

wr=dict()
wrd=dict()
ratio=dict()
ratiod=dict()
degdegout=dict()
degdegoutd=dict()
modularity=dict()
modularityd=dict()
nbcom=dict()
nbcomd=dict()

modu=nk.community.Modularity()

mks=[mk for mk in mks if mk in ["mk100", "mk500"]]
for mk in mks:
    chemin2=os.path.join(chemin1,mk)
    ks=os.listdir(chemin2)
    ks=[k for k in ks if k in ["k15","k25"]]
    for k in ks:
        chemin3=os.path.join(chemin2,k)
        mus=os.listdir(chemin3)
        mus=[mu for mu in mus if mu in ["muw0.2","muw0.4"]]
        for mu in mus:
            chemin4=os.path.join(chemin3,mu)
            ids=os.listdir(chemin4)
            ids=[identif for identif in ids if identif < "20"]
            wr[chemin4]=[]
            wrd[chemin4]=[]
            ratio[chemin4]=[]
            ratiod[chemin4]=[]
            degdegout[chemin4]=[]
            degdegoutd[chemin4]=[]
            modularity[chemin4]=[]
            modularityd[chemin4]=[]
            nbcom[chemin4]=[]
            nbcomd[chemin4]=[]
            for identif in ids:
                chemin=os.path.join(chemin4,identif)
                #TabOne means that the edge list is splitted according to tabs and that the nodes ids starts at 1
                G=nk.graphio.readGraph(os.path.join(chemin, gpath), weighted=True, fileformat=nk.Format.EdgeListTabOne)
                gt_partition=nk.community.readCommunities(os.path.join(chemin, gpath), format="edgelist-t1")
                detected=nk.community.detectCommunities(G)
                wr[chemin4].append(weightingRatio(G,gt_partition))
                wrd[chemin4].append(weightingRatio(G,detected))
                ratio[chemin4].append(ratios(G, gt_partition))
                ratiod[chemin4].append(ratios(G, detected))
                deg,deg_out=node_recall(G, gt_partition) 
                deg=[value for (key, value) in sorted(deg.items())]
                deg_out=[value for (key, value) in sorted(deg_out.items())]
                degdegout[chemin4].append(spearmanr(deg,deg_out ))
                deg,deg_out=node_recall(G, detected) 
                deg=[value for (key, value) in sorted(deg.items())]
                deg_out=[value for (key, value) in sorted(deg_out.items())]
                degdegoutd[chemin4].append(spearmanr(deg,deg_out ))
                modularity[chemin4].append(modu.getQuality(gt_partition,G))
                modularityd[chemin4].append(modu.getQuality(detected,G))
                nbcom[chemin4].append(gt_partition.numberOfSubsets())
                nbcomd[chemin4].append(detected.numberOfSubsets())
                

#chemin2=os.path.join(chemin1,"mk" + args.maxk)
#chemin3=os.path.join(chemin2,"k"+args.avgk)
#chemin4=os.path.join(chemin3,"muw"+args.muw)
```
